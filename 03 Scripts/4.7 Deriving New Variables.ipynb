## Create subset with the first million rows

```python
df = df_orders_products_merged[:1000000]
```

```python
# Check shape

df.shape
```

```python
# Define a function

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid range product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return np.nan
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis = 1)
```

```python
# Check output

df.head()
```

```python
# Check the values

df['price_range'].value_counts(dropna = False)
```

```python
# Check what the most expensive product within subset is

df['prices'].max()
```

## Use the loc() function instead for subset

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

## Use the loc function for the entire df_orders_products_merged dataframe

# Use for-loop to create a 'busiest_day' column

```python
# Find out which day of the week most orders take place

df_orders_products_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# For-loop

result = []

for value in df_orders_products_merged['orders_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

```python
# Print the result list

result
```

```python
# Create a 'busiest_day' column in df_orders_products_merged dataframe

df_orders_products_merged['busiest_day'] = result
```

```python
# Check output

df_orders_products_merged.head()
```

```python
# Check frequency of 'busiest_day' column

df_orders_products_merged['busiest_day'].value_counts(dropna = False)
```

# Task 4.7 step 2: Create a new 'busiest_days' column with 2 busiest days and 2 least busy days

```python
# Use for-loop

result_2 = []

for value in df_orders_products_merged['orders_day_of_week']:
    if value == 0 or value == 1:
        result_2.append('Busiest days')
    elif value == 3 or value == 4:
        result_2.append('Least busy days')
    else:
        result_2.append('Regularly busy')
```

```python
# Print the result_2 list

result_2
```

```python
# Create a 'busiest_days' column in df_orders_products_merged dataframe

df_orders_products_merged['busiest_days'] = result_2
```

```python
# Check output

df_orders_products_merged.head()
```

# Step 3: Check the values of the new 'busiest_days' column

```python
df_orders_products_merged['busiest_days'].value_counts(dropna = False)
```

## The frequency checks out with the 'orders_day_of_week' frequency counts

# Step 4: Create a new 'busiest_period_of_day' column

```python
# Find out frequency for 'order_hour_of_day' column

df_orders_products_merged['order_hour_of_day'].value_counts(dropna = False)
```

## Since there are 3 categories for busiest_period_of_day, each category will contain 8 hours following the previous frequency list

```python
# Use for-loop

result_3 = []

for value in df_orders_products_merged['order_hour_of_day']:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result_3.append('Most orders')
    elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
        result_3.append('Average orders')
    else:
        result_3.append('Fewest orders')
```

```python
# Print the result_3 list

result_3
```

```python
# Create a 'busiest_period_of_day' column in df_orders_products_merged dataframe

df_orders_products_merged['busiest_period_of_day'] = result_3
```

```python
# Check output

df_orders_products_merged.head()
```

# Step 5: Check frequency of 'busiest_period_of_day' column

```python
df_orders_products_merged['busiest_period_of_day'].value_counts(dropna = False)
```
