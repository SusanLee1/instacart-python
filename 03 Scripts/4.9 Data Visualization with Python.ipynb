# Sampling data for line charts

```python
# Create a list holding True/False values to the test np.random.rand() <= 0.7

np.random.seed(4)
dev = np.random.rand(len(df_orders_products_merged)) <= 0.7
```

```python
# Check list

dev
```

```python
# Store 70% of the sample in the dataframe big

big = df_orders_products_merged[dev]
```

```python
# Store 30% of the sample in the dataframe small

small = df_orders_products_merged[~dev]
```

```python
# Check rows of original df_orders_products_merged dataframe

len(df_orders_products_merged)
```

```python
# Check rows of big and small dataframes combined

len(big) + len(small)
```

```python
# Reduce the sample to only necessary columns for line chart

df_2 = small[['orders_day_of_week', 'prices']]
```

# Create a line chart

```python
line = sns.lineplot(data = df_2, x = 'orders_day_of_week', y = 'prices')
```
